```python
import cv2
import copy
import matplotlib.pyplot as plt
import numpy as np
import os
from ultralytics.data.augment import RandomHSV
```

```python
hsv_aug = RandomHSV(hgain=0.01, sgain=0.2, vgain=0.4)
image = {"img": img_rgb.copy()}
img_hsv = hsv_aug(image)
plt.imshow(img_hsv["img"])
```
